```python
orders = pd.read_csv("../Data/orders.csv")
products = pd.read_csv("../Data/products.csv")
sales = pd.read_csv("../Data/sales.csv")
customers = pd.read_csv("../Data/customers.csv")
```

```python
prod = pd.get_dummies(sales["product_id"])
required_columns = list(prod.columns)
required_columns.append("order_id")
sales_updated = pd.concat([sales,prod],axis =1)[required_columns]
sales_updated.head()
```

```python
agg_dict = {i: 'sum' for i in prod.columns}
cart_corr = sales_updated.groupby("order_id").agg(agg_dict).reset_index()[prod.columns]
correlation_matrix = cart_corr.corr()
```

```python
def top_5_products(correlation_matrix,product_ID):
    prod_corr =correlation_matrix[product_ID].values
    top5_indices = np.argsort(prod_corr)[-6:]
    top5_indices = top5_indices[::-1]+1
    return top5_indices[1:]
print(top_5_products(correlation_matrix,1))
```
